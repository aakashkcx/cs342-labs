## Example 2

This second example introduces the use of lists and dictionaries as data structures on Python. These are the building blocks for various Python data structures, including DataFrames, which can be used to store instances
of data (similar to records in a database).

 The following lines of codes exemplify how to work with lists on Python.

````python
myList = [1,2,3,4]


#range() creates a list of numbers from 0 to 9 in increments of 1
myList2 = list(range(0,10,1)) #list of numbers 0 to 9

#this appends the object myList to the end of myList2 - i.e. the last element of
#myList2 is a list
myList2.append(myList)
print (myList2)
````

````python
#remove(x) removes the first element in the list that matches x
myList2.remove(myList)
print (myList2)
````

````python
#to add each element of myList to myList2, we can use a for loop
for x in myList:
    myList2.append(x)
print (myList2)
````

````python
#len(x) returns the length of x
print (len(myList2))
````

````python
#the sort function can take additional parameters such as a key,
#ascending/descending, etc
myList2.sort()
print (myList2)
````

````python
#slicing lists - allows you to select segments of the list
print (myList2[2:4])
print (myList2[:3]) # first 3 elements
````

````python
#dict contains key value pairs
myDict = {'a':'hello','b':'world'}
print (myDict['a'])
print (myDict['b'])
````

### 6.1 Plotting a simple linear function

````python
#We can use plt(imported from matplotlib) to plot a simple linear function
#Defines a linear function with respect to the all Olympic Years in male100.csv
y = -0.014*male100['Year']+38

plt.plot(male100['Year'],y,'r-',color = 'b')
ax = plt.gca() # gca stands for 'get current axis'
ax = male100.plot(x=0,y=1, kind ='scatter', color='g', label="Mens 100m", ax = ax)
female100.plot(x=0,y=1, kind ='scatter', color='r', label="Womens 100m", ax = ax)
````

The above code uses the function *DataFrame.plot(x, y, kind, color, label, ax).* The first two parameters
indicate which columns of the DataFrame are used for the $x$ and $y$ axes. *color* is used to assign a color
to the data points, *marker* defines their shape, and *label* gives the graph a name. *kind* allows the plotting method to be set, the default is *line*, however, many other plotting methods are available, including:

- *bar* or *barh* for bar plots
- *hist* for histogram
- *box* for boxplot
- *kde* or *density* for density plots
- *area* for area plots
- *scatter* for scatter plots
- *hexbin* for hexagonal bin plots
- *pie* for pie plots

A full list of parameters that can be used within *DataFrame.plot()* may be found at http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.html. Additional color options are also available http://matplotlib.org/api/colors_api.html, as are marker shapes
http://matplotlib.org/api/markers_api.html. To plot different DataFrames on the same
figure, we can use the ax variable, for example:
~~~~
    ax = df1.plot()
    df2.plot(ax=ax)
~~~~

## Example 7

Example 7 shows how to read and display a color image in RGB (Red-Green-Blue) format. To this end, we need to load matplotlib library. We also need to import the OpenCV library for easy conversion and image manipulation.

````python
import matplotlib.pyplot as plt
import cv2 # import open CV for python
````

The code below reads the Lena image and displays some basic information about it. 

````python
# Loads an RGB image

# location of image file
#Use the correct path to the file if you are working on your machine 
data_dir = './'

image = plt.imread(data_dir + 'lena.jpg') # lena.jpg is the file name
#image = Image.open(data_dir + 'lena.jpg')
# print type and size of the image (pixels and number of color channels)

print('image type:', image.dtype)
print('image shape:', image.shape)

# The image size is: rows x columns x channels. 
# Recall that color images in RGB format have three color channels: Red, Blue and Green
````

````python
# Displays only a section of the the color image (from row 50 to 200 and from column 50 to 200) 
# using imshow in matplotlib.pyplot library

plt.imshow(image[50:200,50:200,:], )
plt.show()
````

````python
# Prints on the screen the pixel values of each channel and displays the color image (from row 50 to 200 
# and from column 50 to 200)
print ('*'*30 ,'Lena image' ,'*'*30 )
print (image[50:200,50:200,:])
plt.imshow(image[50:200,50:200,:])
plt.show()
````

````python
# Displays the Red channel and prints on the screen its pixel values (from row 50 to 200 and from column 50 to 200)
print ('*'*30 ,'Red channel', '*'*30)
img_R = image[50:200,50:200,0]
print (img_R)

# Use a red color map to display the Red channel
plt.imshow(img_R, cmap='Reds')
plt.show()
````

````python
# Displays the Green channel and prints on the screen its pixel values (from row 50 to 200 and from column 50 to 200)
print ('*'*30 ,'Green channel', '*'*30)
img_G = image[50:200,50:200,1]
print (img_G)

# Use a green color map to display the Green channel
plt.imshow(img_G, cmap='Greens')
plt.show()
````

````python
# Displays the Blue channel and prints on the screen its pixel values (from row 50 to 200 and from column 50 to 200)
print ('*'*30 ,'Blue channel', '*'*30)
img_B = image[50:200,50:200,2]
print (img_B)

# Use a blue color map to display the Blue channel
plt.imshow(img_B, cmap='Blues')
plt.show()
````

It is important to note that the Matplotlib library reads an image in RGB format. However, the OpenCV library reads an image as BGR (Blue, Green, Red). So, if you read an image with Matplotlib library (plt) and you want to use it with OpenCV functions, you need to convert the image from RGB to BGR. The conversion is shown below:

````python
img_BGR = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
````

To convert an image from the BGR color space to the YUV color space, you may use *cv2.cvtColor* with a flag *cv2.COLOR_BGR2YUV*. A list of color space conversions is available at:   https://docs.opencv.org/master/d8/d01/group__imgproc__color__conversions.html

## Example 8

An image is just a matrix of pixel values that can be reshaped to any dimension. For example, to change the shape of an image, we can use the function *reshape()* or if we want to flatten an image, we can use the function *flatten()* or the function *ravel()*. The code below flattens an image into a 1-column vector of dimensions: ((image_height x image_width x no_of_channels), 1).

````python
image = plt.imread(data_dir + 'lena.jpg') # load lena.jpg 
print ('Image dimensions:', image.shape)
reshaped_img1 =  image.ravel() 
print ('Vector dimensions after flattening:' , reshaped_img1.shape)
````

The trailing comma after in (196608**,)** indicates that the result is a tuple. A tuple  is a sequence of $n$ elements.

The *reshape()* function can also be used to flatten an image to a 1-row vector. The argument -1 helps to deduce the dimensions of the vector given the image dimensions.

````python
print ('Image Dimensions:', image.shape)
reshaped_img2 =  image.reshape(1,-1)
reshaped_img3 =  image.reshape(1,image.shape[0] * image.shape[1] * image.shape[2])
print ('Vector dimensions after reshaping (using -1):' , reshaped_img2.shape)
print ('Vector dimensions after reshaping (using actual dimensions):' , reshaped_img3.shape)
````
